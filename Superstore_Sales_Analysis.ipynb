<h3 style = "color:green">4.1 Sales

```python
data['Sales'].describe()
```

```python
data['Sales'].skew()
```

<div style = "color:green; font-size:20px">Since skewness is 13, the sales data is heavily positively skewed. Sales from most of the customers are low to medium, but few sales from very few number of customers is very high. 75% of sales are between 0 to 210 and >210 are from rest 25% of the sales.

```python
bins = np.linspace(data['Sales'].min(), data['Sales'].max(), 50)
plt.hist(data['Sales'], bins = bins, color='lightblue', edgecolor='green')


plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.xticks(np.linspace(data['Sales'].min(), data['Sales'].max(), 20), rotation=45)
plt.title('Distribution of Sales')
plt.show()
```

<div style = "color:green; font-size:20px">From the hist plot, we can observe the sales data is heavily right skewed. Sales from majority of customers are in the range of 0-2000 and sales from only few number of customers are extremely hight.

```python
plt.boxplot(data['Sales'])

plt.title("Box Plot of Sales")
plt.show()
```

```python
labels =['small', 'medium', 'large','enterprise']
bins = [0, 100, 500, 1000, data['Sales'].max()]
data['Sales_Bin'] = pd.cut(data['Sales'], bins = bins, labels = labels)
print(data['Sales_Bin'])


data['Sales_Bin'].value_counts()
data['Sales_Bin'].value_counts().plot(kind = 'bar', color = plt.cm.Greens(np.linspace(0, 1, 4)), edgecolor = 'magenta')
plt.xlabel('Sales Bin')
plt.ylabel('Frequency')
plt.xticks(rotation = 45)
plt.title('Frequency of Sales Bins')
plt.show()

```

<div style = "color:green; font-size:20px">Added sales bins column to divide the sales into small, medium, large and enterprise level categories, since the data is heavily skewed and high values are present as outliers.<br>
Also, more than 75% of the sales are from small and medium categories(0-500).

<div style = "color:green; font-size:25px">4.3 Country, City and State

```python
data['Country'].value_counts()
```

<div style = "color:green; font-size:20px">This store is completely US based.

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))

#For Top cities in the dataset
top_cities = data['City'].value_counts().iloc[:10]
other_cities = data['City'].value_counts().iloc[10:].sum()
top_cities['Others'] = other_cities
top_cities = top_cities.sort_values(ascending = False)

ax[0].bar(top_cities.index, top_cities.values, color = plt.cm.Greens(np.linspace(0.9, 0.1, len(top_cities))), edgecolor = 'magenta')
ax[0].set_xlabel('Cities')
ax[0].set_ylabel('Frequency')
ax[0].set_title('Top 10 Cities')
ax[0].tick_params(axis = 'x', rotation = 45)


#For Top states in the dataset
top_states = data['State'].value_counts().iloc[:10]
other_states = data['State'].value_counts().iloc[10:].sum()
top_states['Others'] = other_states
top_states = top_states.sort_values(ascending = False)

ax[1].bar(top_states.index, top_states.values, color = plt.cm.Greens(np.linspace(0.9, 0.1, len(top_states))), edgecolor = 'magenta')
ax[1].set_xlabel('States')
ax[1].set_ylabel('Frequency')
ax[1].set_title('Top 10 States')
ax[1].tick_params(axis = 'x', rotation = 45)



plt.tight_layout()
plt.subplots_adjust(wspace = 0.4)
plt.show()



```

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))

#For Top cities vs remaining cities
top_cities_count = top_cities.iloc[1:].sum()
remaining_cities_count = top_cities.sum()-top_cities_count
ax[0].pie([top_cities_count, remaining_cities_count], labels = ['Top 10 Cities', 'Remaining Cities'], colors = ['#c6dbef', '#9ecae1'], autopct = '%1.1f%%')
ax[0].set_title('Top 10 Cities vs Remaining Cities')

#For Top states vs remaining states
top_states_count = top_states.iloc[1:].sum()
remaining_states_count = top_states.sum() - top_states_count
ax[1].pie([top_states_count, remaining_states_count], labels = ['Top 10 States', 'Remaining States'], colors = ['#c6dbef', '#9ecae1'], autopct = '%1.1f%%')
ax[1].set_title('Top 10 States vs Remaining States')

plt.legend()
plt.tight_layout()
plt.show()
```

<div style = "color:green; font-size:20px">
<ul>
<li>For Cities: Top 10 cities in the country represent 44% of all cities in the dataset.</li>
<li>For States: Top 10 states in the country represent 70% of all states in the dataset.</li>

<div style = "color:green; font-size:25px">5.4 Region vs Sales

```python

fig, ax = plt.subplots(figsize = (6,6))

#For Outer Ring Sales
sales = data['Sales'].groupby(data['Region']).sum().sort_values(ascending = False)
ax.pie(sales.values, labels = sales.index, colors = plt.cm.Blues(np.linspace(0.8,0.2,4)), autopct = "%1.1f%%", radius = 1,
       textprops = {
           'color':'white'
       })



#For Inner Ring Freq Distribution
region = data['Region'].value_counts().sort_values(ascending = False)
ax.pie(region.values, labels = region.index, colors = plt.cm.Greens(np.linspace(0.8,0.2,4)), autopct = "%1.1f%%", radius = 0.7)

plt.title('Region vs Sales')
plt.show()
```

West and East account for the most in both order volumne and total revenue.Also, East region performs slightly better in terms of value per order, whereas the South region contributes relatively less revenue compared to its order frequency.
